```python
#构建查找表
def creat_lookup_tables(input_data):
    vocab = set(input_data)
    #文字到数字的映射
    vocab_to_int = {word:index for index,word in enumerate(vocab)}
    #数字到文字的映射
    int_to_vocab = dict(enumerate(vocab))
    return vocab_to_int,int_to_vocab
    
```

```python
#首先设置超参

#训练循环次数
nums_epochs = 200

#batch的大小
batch_size = 256

#LSTM层中包含的unit个数
rnn_size = 512

# embedding layer的大小（嵌入层）
embed_dim = 512

# 训练步长
seq_length = 30

# 学习率
learning_rate = 0.003

# 每多少步打印一次训练信息
show_every_n_batches = 30

# 保存session状态的位置
save_dir = 'G:\\save'
```

```python
#inputs和targets的类型都是整数的
def get_inputs():
    inputs = tf.placeholder(tf.int32,[None,None],name='inputs')
    targets = tf.placeholder(tf.int32,[None,None],name='targets')
    learning_rate = tf.placeholder(tf.float32,name='learning_rate')
    return inputs,targets,learning_rate
```

```python
#创建rnn cell，使用lstm cell，并创建相应层数的lstm层，应用dropout，以及初始化lstm层状态。
def get_init_cell(batch_size, rnn_size):
    #lstm层数
    num_layers = 2
    
    # dropout时的保留概率
    keep_prob = 0.8
    
    # 创建包含rnn_size个神经元的lstm cell
    cell = tf.nn.rnn_cell.BasicLSTMCell(rnn_size)
    
    #使用dropout机制防止overfitting等
    drop = tf.nn.rnn_cell.DropoutWrapper(cell,output_keep_prob=keep_prob)
    
    #创建2层lstm层
    cell = tf.nn.rnn_cell.MultiRNNCell([drop for _ in range(num_layers)])
    
    #初始化状态为0.0
    init_state = cell.zero_state(batch_size,tf.float32)
    
    # 使用tf.identify给init_state取个名字，后面生成文字的时候，要使用这个名字来找到缓存的state
    init_state = tf.identity(init_state,name='init_state')
    
    return cell,init_state
```

```python
#创建输入层，提升效率
def get_embed(input_data, vocab_size, embed_dim):
    # 先根据文字数量和embedding layer的size创建tensorflow variable
    embedding = tf.Variable(tf.random_uniform(embed_dim,vocab_size),dtype=tf.float32)
    #让tensorflow帮我们创建lookup table
    return tf.nn.embedding_lookup(embedding,input_data)
```

```python
#创建rnn节点，使用dynamic_rnn方法计算出output和final_state
def build_rnn(cell,inputs):
    '''
    cell就是上面get_init_cell创建的cell
    '''
    outputs, final_state = tf.nn.dynamic_rnn(cell,inputs,dtype=tf.float32)
    # 同样给final_state一个名字，后面要重新获取缓存
    final_state = tf.identity(final_state,name='final_state')
    return outputs,final_state
```

```python
#用上面定义的方法创建rnn网络，并接入最后一层fully_connected layer计算rnn的logits
def build_nn(cell,rnn_size,input_data,vocab_size):
    # 创建embedding layer
    embed = get_embed(input_data,vocab_size,rnn_size)
    # 计算outputs 和 final_state
    outputs,final_state = build_rnn(cell,embed)
    # remember to initialize weights and biases, or the loss will stuck at a very high point
    #记得初始化权重和偏差，否则这个损失将会停留在一个非常高的位置
    logits = tf.layers.dense(outputs,vocab_size)     #简单测试一下，没找到正确的参数设置方法
    return logits,final_state
```

```python
#那么大的数据量不可能一次性都塞到模型里训练，所以用get_batches方法一次使用一部分数据来训练
def get_batch(int_text,batch_size,seq_length):
    #计算有多少个batch可以创建
    num_batchs = ((len(int_text)//(batch_size*seq_length)))
    #计算每一步的原始数据，和位移一位之后的数据
    batch_origin = np.array(int_text[:num_batchs*batch_size*seq_length])
    batch_shifted = np.array(int_text[1:num_batchs*batch_size*seq_length + 1])
    # 将位移之后的数据的最后一位，设置成原始数据的第一位，相当于在做循环
    batch_shifted[-1] = batch_origin[0]
    
    batch_origin_reshape = np.split(batch_origin.reshape(batch_size,-1),num_batchs,1)
    batch_shifted_reshape = np.split(batch_shifted.reshape(batch_size,-1),num_batchs,1)
    
    batchs = np.array(list(zip(batch_origin_reshape,batch_shifted_reshape)))
    return batchs
```

```python
#创建整个RNN网络模型
from tensorflow.contrib import legacy_seq2seq

train_graph = tf.Graph()
with train_graph.as_default():
    #文字总量
    vocab_size = len(int_to_vocab)
    
    # 获取模型的输入，目标以及学习率节点，这些都是tf的placeholder
    input_text, targets, lr = get_inputs()
    
    # 输入数据的shape
    input_text_shape = tf.shape(input_text)
    
    # 创建rnn的cell和初始状态节点，rnn的cell已经包含了lstm，dropout
    # 这里的rnn_size表示每个lstm cell中包含了多少的神经元
    cell, initial_state = get_init_cell(input_text_shape[0],rnn_size)
    
    # 创建计算loss和finalstate的节点
    logits, final_state = build_nn(cell, rnn_size, input_text, vocab_size)
    
    # 使用softmax计算最后的预测概率
    probs = tf.nn.softmax(logits,name='probs')
    
    #计算loss
    cost = legacy_seq2seq.sequence_loss(logits,targets,tf.ones([input_text_shape[0],input_text_shape[1]]))
    
    # 使用Adam梯度下降
    optimizer = tf.train.AdamOptimizer(lr)
    
    # 裁剪一下Gradient输出，最后的gradient都在[-1, 1]的范围内
    gradients = optimizer.compute_gradients(cost)
    capped_gradients = [tf.clip_by_value(grad,-1.,1.)for grad,var in gradients if grad is not None]
    train_op = optimizer.apply_gradients(capped_gradients)
```
